## 3.2 벡터

### 3.2.1 벡터의 정의

벡터는 스칼라의 집합이며, 행렬을 구성하는 기본 단위  
벡터는 크기와 방향을 모두 나타내는 개념  
벡터를 구성하는 각 숫자는 공간상의 좌표를 나타냄  
화살표의 길이가 크기에 해당, 화살표가 향하는 방향이 벡터의 방향을 의미  
특별한 언급이 없으면 벡터는 기본적으로 열 벡터를 의미  
*Reference)* 데이터셋을 하나의 행렬이라고 생각하면 행 벡터는 하나의 **데이터 레코드**라고 생각할 수 있고, 열 벡터는 특정 **피처** 값들의 모음이라 할 수 있다.  

### 3.2.2 벡터의 덧셈과 뺄셈

* 벡터의 덧셈: 더하고자 하는 두 벡터의 크기가 동일할 때만 연산이 가능, 교환 법칙이 성립
* 벡터의 뺄셈: 덧셈의 연장선

### 3.2.3 벡터의 스칼라 곱

벡터에 1보다 작은 양수를 곱하면 해당 벡터는 기존 벡터보다 작아지고,  
음수를 곱하면 크기가 바뀔 뿐만 아니라 방향도 정반대로 바뀐다.

### 3.2.4 벡터 기본 연산의 성질

벡터의 덧셈: 교환 법칙, 결합 법칙이 성립

### 3.2.5 파이썬 실습

#### 벡터의 덧셈

```python
u = [1, 2, 3]
v = [4, 5, 6]

w = []

```

* 함수로 만들기

````python
def v_add(u, v):
    """
    벡터의 덧셈
    입력값: 더하고자 하는 벡터 u, v
    출력값: 벡터 u, v의 덧셈 결과 w
    """
    n = len(u)
    w = []

    for i in range(0, n):
        val = u[i] + v[i]
        w.append(val)
    
    return w
````

````python
# 벡터의 뺄셈
def v_subtract(u, v):
    """
    벡터의 뺄셈
    입력값: 빼고자 하는 벡터 리스트 u, v
    출력값: 벡터 u, v의 뺄셈 결과 w
    """
    n = len(u)
    w = []

    for i in range(0, n):
        val = u[i] - v[i]
        w.append(val)
    return w

# 스칼라와 벡터의 곱
# 특정 스칼라 값이 벡터를 구성하는 각 원소에 모두 곱해지는 것을 의미  
def scalar_v_mul(a, u):
    """
    벡터의 스칼라 곱
    입력값: 스칼라 a, 벡터 리스트 u
    출력값: 스칼라 a와 벡터 리스트 u의 곱 결과 w
    """

    n = len(u)
    w = []

    for i in range(0, n):
        val = a * u[i]
        w.append(val)

    return w

# 벡터의 원소 곱
# 두 벡터의 동일한 위치의 원소를 곱하는 연산
def v_mul(u, v):
    """
    벡터의 원소 곱
    입력값: 원소를 곱하고자 하는 벡터 리스트 u, v
    출력값: 벡터 u, v의 원소 곱 결과 w
    """
    n = len(u)
    w = []

    for i in range(0, n):
        val = u[i] * v[i]
        w .append(val)

    return w

# 벡터의 나눗셈
# 동일한 위치의 원소끼리 나누는 것
def v_div(u, v):
    """
    벡터의 원소 나눗셈
    입력값: 원소를 나눗셈하고자 하는 벡터 리스트 u, v
    출력값: 벡터 u, v의 원소 나눗셈 결과 w
    """
    n = len(u)
    w = []
    
    for i in range(0, n):
        val = u[i] / v[i]
        w.append(val)

    return w
````

### 3.3.1 행렬의 정의

### 3.3.2 행렬의 덧셈과 뺄셈

각 행렬의 동일 위치에 대응하는 원소끼리 덧셈 혹은 뺄셈을 함  
더하거나 빼려는 각 행렬의 크기가 동일하다는 전제가 필요  

### 3.3.3 행렬의 스칼라 곱

*Ref)* 행렬의 곱에는 크게 스칼라 곱, 원소 곱, 행렬 곱이 존재  
행렬에 스칼라를 곱하면 행렬에 포함된 모든 원소에 스칼라를 곱한다.  
기존 행렬에 스칼라 2를 곱한다는 것은 각 행 벡터의 길이가 2배로 늘어난다는 것을 의미  

### 3.3.4 행렬의 원소 곱

동일한 위치의 원소들끼리 서로 곱하는 것을 의미  
행렬 $\mathbf{A}$와 행렬 $\mathbf{B}$의 원소 곱은 $\mathbf{A}\odot\mathbf{B}$라고 표기  

### 3.3.5 행렬 곱

행렬끼리 서로 곱하는 것을 의미  
전제: 곱하는 행렬의 열 크기와 곱해지는 행렬의 행 크기가 일치해야 함.  
행렬 곱 연산은 교환 법칙이 성립하지 않음.  
1) 전제에 따라 자체가 성립하지 않을 수도 있고,
2) 성립은 하지만 행렬의 크기가 다를 수도 있고,
3) 행렬의 구성 성분 값이 다를 수 있다.

### 3.3.6 행렬의 대각합

행렬 $\mathbf{A}$가 정사각 행렬일 때, 행렬 $\mathbf{A}$의 **대각합** (trace) 은 주 대각 원소를 모두 더한 값을 의미  
대각합은 tr($\mathbf{A}$)로 표기  
정사각행렬이 아닌 경우 대각합은 정의되지 않음.

### 3.3.7 행렬 연산의 성질

행렬의 덧셈은 교환 법칙이 성립  
행렬의 덧셈은 결합 법칙이 성립  
행렬 곱은 결합 법칙이 성립  
행렬 연산에서 분배 법칙이 성립  
행렬 곱과 스칼라 곱이 혼용될 때도 분배 법칙이 성립

````python
# 행렬의 덧셈
# 행렬의 덧셈 결과를 구하는 함수 구현  
def add(A, B):
    """
    행렬의 덧셈
    입력값: 행렬의 덧셈을 수행할 행렬 A, B
    출력값: 행렬 A와 행렬 B의 덧셈 결과인 행렬 C
    """
    
    n = len(A) # 행 크기
    p = len(A[0]) # 열 크기
    
    res = []
    for i in range(0, n):
        row = []
        for j in range(0, p):
            val = A[i][j] + B[i][j]
            row.append(val)
        res.append(row)
    return res

# 행렬의 뺄셈
def subtract(A, B):
    """
    행렬의 뺄셈
    입력값: 행렬의 뺄셈을 수행할 행렬 A, B
    출력값: 행렬 A와 행렬 B의 뺄셈 결과인 행렬 C
    """
    
    n = len(A)
    p = len(A[0])
    
    res = []
    for i in range(0, n):
        row = []
        for j in range(0, p):
            val = A[i][j] - B[i][j]
            row.append(val)
        res.append(row)
    return res

# 행렬의 스칼라 곱
def scalar_mul(b, A):
    """
    행렬의 스칼라 곱
    입력값: 스칼라 곱을 수행할 스칼라 b, 행렬 A
    출력값: 스칼라 b와 행렬 A의 스칼라 곱 결과인 행렬 C
    """
    
    n = len(A)
    p = len(A[0])
    
    res = []
    for i in range(0, n):
        row = []
        for j in range(0, p):
            val = b * A[i][j]
            row.append(val)
        res.append(row)
    return res

# 행렬의 원소 곱
# 곱하고자 하는 행렬의 동일한 위치의 원소끼리 곱하는 것을 의미
def ele_product(A, B):
    """
    행렬의 원소 곱
    입력값: 행렬의 원소 곱을 수행할 행렬 A, B
    출력값: 행렬 A와 행렬 B의 원소 곱 결과인 행렬 C
    """
    
    n = len(A)
    p = len(A[0])
    
    res = []
    for i in range(0, n):
        row = []
        for j in range(0, p):
            val = A[i][j] * B[i][j]
            row.append(val)
        res.append(row)
    return res

# 행렬 곱
def matmul(A, B):
    """
    행렬의 행렬 곱
    입력값: 행렬 곱을 수행할 행렬 A, B
    출력값: 행렬 A와 행렬 B의 행렬 곱 결과인 행렬 res
    """
    
    n = len(A)
    p1 = len(A[0])
    p2 = len(B[0])
    
    res = []
    for i in range(0, n):
        row = []
        for j in range(0, p2):
            val = 0
            for k in range(0, p1):
                val += A[i][k] * B[k][j]
            row.append(val)
        res.append(row)
    return res
````
