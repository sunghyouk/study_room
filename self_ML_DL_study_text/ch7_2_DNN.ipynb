```python
# fashione MNIST 불러오기
from tensorflow import keras
(train_input, train_target), (test_input, test_target) = \
keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split
train_scaled = train_input / 255.0 # regularization
# train_scaled = train_scaled.reshape(-1, 28*28) # 가로, 세로 2차원을 1차원으로 만들기
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

## 1st step)
keras의 dense 클래스 이용  
은닉층: 시그모이드 활성화 함수  
출력층: 소프트맥스 함수  

## 2nd step)
Sequential 클래스에 추가하여 심층 신경망 만들기  
1+2 step 한꺼번에 Sequential에 이름 정해서 넣기

```python
# model = keras.Sequential([
#                          keras.layers.Dense(100, activation='sigmoid', input_shape=(784, ),
#                                             name='hidden'),
#                          keras.layers.Dense(10, activation='softmax', name='output')
# ], name='fashion MNIST model')
```

```python
# add method로 바꾸면 조건에 따라 층을 추가하거나 길어진 코드를 보기 쉬워진다.
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

## 3rd step)
compile하고 fitting하기

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
model.fit(train_scaled, train_target, epochs=5)
```

## 4th step)
validating

```python
model.evaluate(val_scaled, val_target)
```
